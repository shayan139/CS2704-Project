```python
#display a doughnut chart of percentages of weather conditions
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Weather': [
        'Mainly Clear','Mostly Cloudy','Cloudy','Clear','Snow',
        'Rain','Rain Showers','Fog','Drizzle,Fog','Rain,Fog',
        'Snow Showers','Drizzle','Snow,Fog'
    ],
    'Frequency': [2106,2069,1728,1326,390,306,188,150,80,116,60,42,37]
})

# keep big slices, group rest into “Other”
series = df.set_index('Weather')['Frequency']
big = series[series/series.sum() >= 0.03].copy()
big['Other'] = series[series/series.sum() < 0.03].sum()

fig, ax = plt.subplots(figsize=(9, 7))

w = 0.3                              # donut thickness
center = 1 - w/2                     # radial center of the ring

wedges, texts, autotexts = ax.pie(
    big,
    labels=None,                     # no slice labels
    autopct='%1.1f%%',               # just percentages
    startangle=90,
    pctdistance=center,              # ← move % to ring midpoint
    wedgeprops=dict(width=w),        # ← make it a donut
    textprops={'ha':'center','va':'center','fontsize':10}
)

ax.legend(
    wedges,
    big.index,
    title='Weather',
    loc='center left',
    bbox_to_anchor=(1, 0.5),
    fontsize='small'
)

ax.set_title('Weather Conditions')
ax.axis('equal')
plt.tight_layout()
plt.show()
```
